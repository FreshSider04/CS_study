1. 3개의 임의의 정수 x, y, z를 입력받아, 세 수 중 가장 큰 수를 출력하는 프로그램을 작성하라.
   - x, y, z는 임의의 정수(음수, 0, 양수 모두 가능)
   - min, max 함수를 사용하지 않고 if 문장으로만 가장 큰 수를 결정

```python
# 1
x, y, z = map(int, input().split())

if x > y:
    max_number = x
else:
    max_number = y


if max_number < z:
    max_number = z
    
print("입력한 수 :", end=" ")
print(x, y, z)
max_number
```

2. 십진수 정수 안의 각 자리수의 합을 계산하는 프로그램을 작성하라. 예를 들어 2489 의 각 자리수의 합은 (2+4+8+9) = 23 임
   - 십진 정수를 입력 받음(자리수는 정해지지 않음)
   - 입력받은 십진수가 음수이면 음수 출력 (예, -12) => -3

```python
# 2
number = input("10진수의 정수를 입력해 주세요 : ")
sum_number = 0

if number[0] != '-':
    for i in range(0, len(number)):
        sum_number += int(number[i])

else:
    for i in range(1, len(number)):
        sum_number -= int(number[i])

print(sum_number)
```

```python
# 3 (4)

import re
from collections import Counter

def replaceSpace(text):
    return re.sub(r"[,\.“”‘’()~\-…·]", " ", text)

with open('news.txt', 'r', encoding='cp949') as file:
    text = file.read()
    text = replaceSpace(text)
    
    words = text.split()

    wordCount = Counter(words)
    sorted_word_counts = sorted(wordCount.items(), key=lambda x: x[1], reverse=True)
    
    for word, count in sorted_word_counts:
        print(f"{word}: {count}")
    
    file.close()
```

4. 임의의 실수(10진수)를 입력 받아 임의의 진수(16진 이하)로 변환하는 함수 to_decimal_to_binary를 프로그램하라.   
   - 실수와 변환할 진수는 입력 받음   
   - 11~15진수의 경우 10은 A로, 11은 B로, 12는 C로, 13은 D로, 14은 E로, 15은 F로 한자리 수로 표시하라.   
   예) 13.125(10) => D.2(16)   

```python
# 4
def to_decimal_to_binary():
    # 입력 받기
    num = float(input("변환할 실수를 입력하세요: "))
    base = int(input("변환할 진수를 입력하세요 (2 ~ 16): "))
    
    # 유효한 진수인지 확인
    if base < 2 or base > 16:
        print("유효한 진수를 입력하세요 (2 ~ 16)")
        return
    
    # 정수 부분과 소수 부분 분리
    int_part = int(num)
    float_part = num - int_part
    
    # 정수 부분 변환
    int_result = ''
    if int_part == 0:
        int_result = '0'
    else:
        while int_part > 0:
            remainder = int_part % base
            if remainder >= 10:
                int_result = chr(55 + remainder) + int_result
            else:
                int_result = str(remainder) + int_result
            int_part //= base
    
    # 소수 부분 변환
    float_result = ''
    precision = 10  # 소수점 이하 10자리까지 계산
    while float_part > 0 and len(float_result) < precision:
        float_part *= base
        digit = int(float_part)
        if digit >= 10:
            float_result += chr(55 + digit)
        else:
            float_result += str(digit)
        float_part -= digit
    
    # 결과 출력
    if float_result == '':
        print(f"{num}(10) => {int_result}({base})")
    else:
        print(f"{num}(10) => {int_result}.{float_result}({base})")

# 함수 호출
to_decimal_to_binary()
```

5. list comprehension & 표준라이브러리 random 활용 문제
   (1) list comprehension을 이용하여 [1, 1000] (1, 1000 포함) 사이의 랜덤 정수 100개를 가지는 리스트 randomA, randomB를 생성하라. (20) 
   (2) 두 randomA와 randomB 리스트의 같은 index의 element를 비교하여 randomA의 element가 큰 경우의 개수를 구하는 list comprehension 프로그램을 작성하라.

```python
# 5
from random import *

count = 0

randomA = [randint(1, 1000) for _ in range(100)]
randomB = [randint(1, 1000) for _ in range(100)]

for i in range(0, 100):
    if randomA[i] > randomB[i]:
        count += 1

count
```

## 격자 그리기 (10 column X 10 row)

```python
import pygame #파이 게임 모듈 임포트

pygame.init() #파이 게임 초기화
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT)) #화면 크기 설정
clock = pygame.time.Clock() 

#변수

BLACK = (0, 0, 0)
BLUE = (0, 0, 255)
WHITE = (255, 255, 255)
CELL_SIZE = 50
COLUMN_COUNT = 10
ROW_COUNT = 10

while True: #게임 루프
    screen.fill(BLACK) #단색으로 채워 화면 지우기

    #변수 업데이트

    event = pygame.event.poll() #이벤트 처리
    if event.type == pygame.QUIT:
        break

    #화면 그리기

    for column_index in range(COLUMN_COUNT):
        for row_index in range(ROW_COUNT):
            pygame.draw.rect(screen, BLUE, pygame.Rect(column_index * CELL_SIZE , row_index * CELL_SIZE , CELL_SIZE, CELL_SIZE)) #커버
            pygame.draw.rect(screen, WHITE, pygame.Rect(column_index * CELL_SIZE, row_index * CELL_SIZE, CELL_SIZE, CELL_SIZE), 1)

    pygame.display.update() #모든 화면 그리기 업데이트
    clock.tick(30) #30 FPS (초당 프레임 수) 를 위한 딜레이 추가, 딜레이 시간이 아닌 목표로 하는 FPS 값

pygame.quit() 
```

## 타일 열기

```python
import pygame #파이 게임 모듈 임포트
import random

def in_bound(column_index, row_index):
    if (0 <= column_index < COLUMN_COUNT and 0 <= row_index < ROW_COUNT):
        return True
    else:
        return False

def open_tile(column_index, row_index): 
    if not in_bound(column_index, row_index):
        return

    tile = grid[row_index][column_index]
    if not tile['open']:
        tile['open'] = True
    else:    
        return

pygame.init() #파이 게임 초기화
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT)) #화면 크기 설정
clock = pygame.time.Clock() 

#변수

BLACK = (0, 0, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
WHITE = (255, 255, 255)
small_font = pygame.font.SysFont('malgungothic', 36)
CELL_SIZE = 50
COLUMN_COUNT = SCREEN_WIDTH // CELL_SIZE
ROW_COUNT = SCREEN_HEIGHT // CELL_SIZE

grid = [[{'mine': False, 'open': False} for _ in range(COLUMN_COUNT)] for _ in range(ROW_COUNT)]
MINE_COUNT = 15
for _ in range(MINE_COUNT):
    while True:
        column_index = random.randint(0, COLUMN_COUNT - 1)
        row_index = random.randint(0, ROW_COUNT - 1)
        tile = grid[row_index][column_index]
        if not tile['mine']:
            tile['mine'] = True #지뢰 설치
            break

while True: #게임 루프
    screen.fill(BLACK) #단색으로 채워 화면 지우기

    #변수 업데이트

    event = pygame.event.poll() #이벤트 처리
    if event.type == pygame.QUIT:
        break
    elif event.type == pygame.MOUSEBUTTONDOWN:
        if event.button == 1:
            column_index = event.pos[0] // CELL_SIZE
            row_index = event.pos[1] // CELL_SIZE
            if in_bound(column_index, row_index):
                print(column_index, row_index)
                open_tile(column_index, row_index)

    #화면 그리기

    for column_index in range(COLUMN_COUNT):
        for row_index in range(ROW_COUNT):
            tile = grid[row_index][column_index]
            if tile['mine']: 
                mine_image = small_font.render('x', True, RED)
                screen.blit(mine_image, (column_index * CELL_SIZE + 10, row_index * CELL_SIZE + 10)) #지뢰 설치
            if not tile['open']:
                pygame.draw.rect(screen, BLUE, pygame.Rect(column_index * CELL_SIZE, row_index * CELL_SIZE, CELL_SIZE, CELL_SIZE)) #커버
            if tile['mine']: 
                dot_image = small_font.render('.', True, WHITE) 
                screen.blit(dot_image, (column_index * CELL_SIZE + 10, row_index * CELL_SIZE + 10)) #임시 지뢰 표시
            pygame.draw.rect(screen, WHITE, pygame.Rect(column_index * CELL_SIZE, row_index * CELL_SIZE, CELL_SIZE, CELL_SIZE), 1)

    pygame.display.update() #모든 화면 그리기 업데이트
    clock.tick(30) #30 FPS (초당 프레임 수) 를 위한 딜레이 추가, 딜레이 시간이 아닌 목표로 하는 FPS 값

pygame.quit()
```
